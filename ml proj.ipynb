```python
# NOW WE ARE MOVING TOWARDS THE MACHINE LEARNING PROCESSING

#first of all we know thta we dont have any null values so that thng is good to go
# next is the catergorical varibales
# we need to do feature enginnering on those varubales so that the machine learnign models undertsanf everything perfectly


# we need to do workign on 
# sex
# region adn 
# smokers    we will be doing one hot encoding or label encoding for them

```

```python
# for male and female

med_ins.replace({'sex': {'male': 0, 'female': 1}}, inplace=True)
```

```python
# for smoker and non smoker

med_ins.replace({'smoker': {'yes': 0, 'no': 1}}, inplace=True)
```

```python
# for regions we have four unique values so we will be doing labvelling for that

med_ins.replace({'region': {'southeast':0 , 'southwest':1 , 'northeast':2 , 'northwest':3}}, inplace=True)
```

```python
med_ins.head()   # so encoding is completed
```

```python
# out target is CHARGES
# and the fetaure son the basis of which we will predict the target charges are
# age
# sex
# bmi
# children
# smoker
# region
```

```python
# so for storing the values in x and y we will eb addign all the independant variables in x and dependant in y
```

```python
# this is where we are storing our dependant and indepenedant variales

x = med_ins.drop(columns='charges', axis=1)
y = med_ins['charges']
```

```python
x
```

```python
y
```

```python
# now we will be splitting the datset in training and testing data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
```

```python
# now its time for model training
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
# next is model evaluation (prediction on training data))
```

```python
y_predict = lr.predict(x_test)
y_predict2= lr.predict(x_train)
```

```python
y_predict
```

```python
y_predict2
```

```python
from sklearn.metrics import r2_score, mean_squared_error, confusion_matrix, classification_report
```

```python
r2_test = r2_score(y_test, y_predict)
r2_train = r2_score(y_train, y_predict2)

# Calculate Mean Squared Error
mse_test = mean_squared_error(y_test, y_predict)
mse_train = mean_squared_error(y_train, y_predict2)

print('mse_test:', mse_test)
print('R2_test:', r2_test)

rmse_test=np.sqrt(mse_test)
print('RMSE_test:', rmse_test)
print('-----')
print('mse_train:', mse_train)
print('R2_train:', r2_train)

rmse_train=np.sqrt(mse_train)
print('RMSE_train:', rmse_train)
```

R is high (0.78 means 78% for testting data adn 74% for trainning data)--strong predictive power

MSE looks large but normal for this type of dataset

RMSE= 5800, adn 6110---expected error margin in medical cost prediction

```python
import gradio as gr
import joblib
import numpy as np

# Load your trained model
model = joblib.load(r"C:\Users\areeb\Desktop\S5\ML\project\model.pkl")

# Prediction function
def predict_charges(age, sex, bmi, children, smoker, region):
    features = np.array([[age, sex, bmi, children, smoker, region]])
    prediction = model.predict(features)
    return f"Predicted Medical Insurance Cost: {prediction[0]}"

# Gradio interface
inputs = [
    gr.Number(label="Age"),
    gr.Number(label="Sex (1 = female, 0 = male)"),
    gr.Number(label="BMI"),
    gr.Number(label="Children"),
    gr.Number(label="Smoker (1 = No, 0 = Yes)"),
    gr.Number(label="Region (southeast:0 , southwest:1 , northeast:2 , northwest:3)"), 
]

output = gr.Textbox(label="Predicted Charges")

gr.Interface(
    fn=predict_charges, 
    inputs=inputs, 
    outputs=output,
    title="Medical Insurance Charges Predictor"
).launch()
```
